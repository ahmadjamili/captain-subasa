### most percentage and least percentage

```python
goal_dataset=dataset[dataset['outcome']=='گُل']
```

```python
not_goal_dataset=dataset[dataset['outcome']!='گُل']
```

```python
not_goal_dataset
```

```python
goal_dataset1=goal_dataset.groupby('playerId')[['outcome']].count()
```

```python
goal_dataset1=goal_dataset1.reset_index()
```

```python
goal_dataset1=goal_dataset1.sort_values(by=['outcome'])
```

```python
goal_dataset1
```

```python
not_goal_dataset1=not_goal_dataset.groupby('playerId')[['outcome']].count()
not_goal_dataset1=not_goal_dataset1.reset_index()
not_goal_dataset1=not_goal_dataset1.sort_values(by=['outcome'])
```

```python
goal_dataset1['goal_rate']=goal_dataset1['outcome']/(not_goal_dataset1['outcome']+goal_dataset1['outcome'])
```

```python
goal_dataset1.sort_values(by=['goal_rate'])
```
